```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%load_ext autoreload

%autoreload 2

%matplotlib notebook
sns.set()
```

```python
import galsim
from mdetsims import Sim
```

```python
sim = Sim(
    rng=np.random.RandomState(seed=100),
    g1=0.02,
    gal_type='exp',
    scale=0.263,
    psf_type='gauss',
    n_coadd=30,
    mask_and_interp=True)
```

```python
mbobs = sim.get_mbobs()
```

```python
plt.figure(figsize=(8, 8))
ax = plt.gca()
ax.imshow(np.arcsinh(mbobs[0][0].image))
ax.grid(False)

plt.figure(figsize=(8, 8))
ax = plt.gca()
ax.imshow(np.arcsinh(mbobs[0][0].noise))
ax.grid(False)

plt.figure(figsize=(8, 8))
ax = plt.gca()
ax.imshow(np.arcsinh(mbobs[0][0].bmask))
ax.grid(False)
```
