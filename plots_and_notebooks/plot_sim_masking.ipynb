```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%load_ext autoreload

%autoreload 2

%matplotlib notebook
sns.set()
```

```python
import galsim
from mdetsims import Sim
```

```python
sim = Sim(
    rng=np.random.RandomState(seed=100),
    g1=0.02,
    gal_type='exp',
    scale=0.263,
    psf_type='gauss',
    n_coadd=30,
    mask_and_interp=True,
    add_bad_columns=False)
```

```python
mbobs = sim.get_mbobs()

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].imshow(np.arcsinh(mbobs[0][0].image/np.std(mbobs[0][0].noise)))
axs[0, 0].grid(False)

axs[0, 1].imshow(np.arcsinh(mbobs[0][0].noise/np.std(mbobs[0][0].noise)))
axs[0, 1].grid(False)

axs[1, 0].imshow(np.arcsinh(mbobs[0][0].bmask))
axs[1, 0].grid(False)

axs[1, 1].axis('off')
```
